```
def clear_para(para):
    para = para.replace('<p>', '')
    para = para.replace('</p>', '')
    para = para.replace('<head>', '')
    para = para.replace('</head>', '')
    return para.strip()
```

```
print('=' * 50)
print('21세기 세종계획 원시말뭉치 내용 추출')
print('=' * 50)
```

```
base_path = "/content/gdrive/My Drive/Colab Notebooks/colab_letuer/20200724/"
in_fname = input('input name : ')
out_fname = input('output name : ')

with open(base_path+in_fname,encoding='utf-16') as f:
      paras = [clear_para(para) for para in f if '<p>' in para or '<head>' in para]
      
with open(base_path+out_fname, 'w', encoding='utf-8') as f:
      f.write('\n'.join(paras))

word_cnt = 0
for para in paras:
    word_cnt += para.count(' ')
    word_cnt += 1

print('=' * 50)
print('%d 문단, %d 토큰 추출 완료' % (len(paras), word_cnt))
print('=' * 50)
```
