```
import tqdm
from konlpy.tag import Kkma
from google.colab import drive
```

```
base_path = "/content/"
```

```
# =============================================================================
# 파일 입력
# =============================================================================
try:
    print('=' * 50)
    print('형태소 분석 프로그램')
    print('=' * 50)

    in_fname = input('input name : ')
    with open(base_path+in_fname, encoding='utf-8') as f:
        lines = [line.strip() for line in f]
        
    kkma = Kkma()
    
    print('=' * 50)
    print('분석 파일 생성 방법')
    print('=' * 50)    
    print('1. 모든 형태소 출력 - 품사 정보 미포함')
    print('2. 모든 형태소 출력 - 품사 정보 포함')
    print('3. 명사만 출력')
    print('=' * 50)    
    menu = int(input('>> '))
    
    while True:
        if menu == 1:
            with open('tagging.all.txt', 'w', encoding='utf-8') as f:    
                cnt = 0
                for i in tqdm.trange(len(lines)):
                    f.write('%s\n' % ' '.join(kkma.morphs(lines[i])))
                    
        elif menu == 2:
            with open('tagging.all.pos.txt', 'w', encoding='utf-8') as f:        
                for i in tqdm.trange(len(lines)):
                    morp_list = kkma.pos(lines[i])
                    temp = ['/'.join(morp) for morp in morp_list]                    
                    f.write('%s\n' % ' '.join(temp))                            
        elif menu == 3:
            with open(base_path+'tagging.noun.txt', 'w', encoding='utf-8') as f:        
                for i in tqdm.trange(len(lines)):
                    noun_list = kkma.nouns(lines[i])
                    if len(noun_list) != 0:
                        f.write('%s\n' % ' '.join(noun_list))                            
        else:
            print('메뉴 선택은 숫자만 입력하셔야 합니다.')                            
            menu = int(input('>> '))        
            continue
        break
        
except FileNotFoundError:
    print('해당 파일을 찾을 수 없습니다.')
except UnicodeDecodeError:
    print('문서 인코딩을 확인하세요. utf-8 문서만 처리 가능합니다.')
```
