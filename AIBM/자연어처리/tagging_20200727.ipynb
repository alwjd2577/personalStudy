```
# 한국어 품사 태깅
# !pip install konlpy
from konlpy.tag import Hannanum
hannanum = Hannanum()
sample_text = '이 노트는 비공개 출력으로 열려 있습니다. 출력 내용은 저장되지 않습니다.'
tagging_result = hannanum.pos(sample_text)
print(tagging_result)
print(hannanum.tagset)    # tag에 대한 정보 확인

from konlpy.tag import Kkma
kkma = Kkma()
print(kkma.tagset)

from konlpy.tag import Komoran
komoran = Komoran()
print('\nKomorarn\n', komoran.tagset, '\n')

from konlpy.tag import Okt
okt = Okt()
print(okt.tagset)
```

```
# 원하는 태그만 추출하기

target_han_tag = ['N', 'P', 'X']
filtered_tokens = []

for token, tag in tagging_result:
  if tag in target_han_tag:
    filtered_tokens.append(token)

print(filtered_tokens)
```

```
# Komoran으로 태그 추출해보기
Komorantagging_result = komoran.pos(sample_text)
print('[Komoran Pos Result]\n', Komorantagging_result, '\n')

target_komoran = ['NNP', 'JKB']
Komoran_filtered = []

for token, tag in Komorantagging_result:
  if tag in target_komoran:
    Komoran_filtered.append(token)

print('Komoran_filtered\n',Komoran_filtered)
```
