```
# JSON: 파일로 저장하는 Ditionary???
import json

a = {'value1': 5, 'value2': 10, 'seq':[1, 2, 3, 4]}

filename = 'test.json'
with open(filename, 'w') as f:      # 'w': 쓰기모드, filename을 읽으면 f라고 지칭
    json.dump(a, f)                 # file f에다가 a의 데이터를 덮어씌어라

# with: 구문이 끝나면 자동으로 open을 close해줌
with open(filename, 'r') as f:      # 'r': 읽기 모드, file f를 result로 로드하기
    result = json.load(f)
    print(result)

# 실험결과당 JSON 파일 생성
```
